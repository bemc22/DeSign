```python
#@title Experiment Configuration Setup


enable_threshold = True #@param {type:"boolean"}
enable_center = True #@param {type:"boolean"}
enable_scale = True #@param {type:"boolean"}
resume_training = False #@param {type:"boolean"}

results_path = f'./result_thres={enable_threshold}_center={enable_center}_scale={enable_scale}'

threshold_opt = "--threshold"
center_opt = "--batch_center"
scale_opt = "--batch_scale"

if enable_threshold == False:
  threshold_opt = "--no-threshold"

if enable_center == False:
  center_opt = "--no-batch_center"

if enable_scale == False:
  scale_opt = "--no-batch_scale"

variable_name = 'result_path'
```

```python
#@title run training procedure

run = "python  main.py " + \
"--gpus 0 " + \
"--model resnet18_1w1a " + \
f"--results_dir {results_path} " + \
"--data_path /data " + \
"--dataset cifar10 " + \
"--epochs 3 " + \
"--lr 0.1 " + \
"-b 256 " + \
"-bt 128 " + \
"--lr_type cos " + \
"--warm_up " + \
"--weight_decay 5e-4 " + \
"--tau_min 0.85 " + \
"--tau_max 0.99 " + \
f"{threshold_opt} " + \
f"{center_opt} " + \
f"{scale_opt} "

if resume_training:
  run += "--resume"

!{run}
```
